```python
import pandas 
f = open('crash_facts_16.txt', 'r')
lines = f.readlines()
f.close()
print(lines)
```

As we can see, the car crash data for the year 2017 is not very clearly formated and will take some work in order to extract useful information. We want to convert this txt file into a csv which can be more easily manipulated. In order to do this we will have to add a comma after every column value in every row, in order to do this we have to get rid of the existing commas first.  


```python
lines = [w.replace(',', '') for w in lines]
print(lines)
```

As we can see we have now removed all the uneeded commas and will now add all the commas we do need. 

```python
lines = [w.replace('County', 'County,') for w in lines]
lines = [w.replace('City', 'City,') for w in lines]
lines = [w.replace('Charles City, County,', 'Charles City County,') for w in lines]
lines = [w.replace('James City, County,', 'James City County,') for w in lines]
lines = [w.replace('\n', '') for w in lines]

lines

    
```

```python
def is_number(s):
    try:
        float(s)
        return True
    except ValueError:
        pass
 
    try:
        import unicodedata
        unicodedata.numeric(s)
        return True
    except (TypeError, ValueError):
        pass
 
    return False



```

```python
lines2 = []

lines = list(filter(None, lines))
for line in lines:
    
    line2 = []
    countyPart, numbPart= line.split(',')
    print(countyPart)
    print(numbPart)
    line2.append(countyPart)
    for numb in numbPart.split():
        if (is_number(numb)):
            numb = numb 
        line2.append(numb)
    lines2.append(line2)
```

```python
print(lines2)
```

We now have a list of lists, in which each individual list contains all the column values for the given row. We now want to get this data into a pandas Dataframe which we can then export to a csv. 

```python
df2017 = pandas.DataFrame(lines2)
df2017
```

```python
df2017.columns = ['County', 'Crashes','Fatalities', 'Injuries','Licensed Drivers', 'Death Rate per 1000 Drivers']
df2017
```

We can now export this file to a csv. 

```python
df2017.to_csv("2016clean1.csv")
```
