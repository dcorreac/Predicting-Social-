```python
import pandas
import numpy
import requests
import bs4
import nose.tools
```

```python
vaData = pandas.read_csv('vaData3.csv')
vaData
```

```python
counties = []
for county in vaData.County:
    county = county.split(',')[0].lower()
    counties.append(county)
```

```python
countyRain = []
countySnow = []
countyElevation = []
countyNames = []





for county in counties:
    url = 'https://www.bestplaces.net/climate/county/virginia/' + county
    rain = -1
    snow = -1 
    elevation = -1
    page = requests.get(url)
    print(county)
    f = open('downloaded.html','w')
    f.write(page.text)
    f.close()
    soup = bs4.BeautifulSoup(open("downloaded.html"), 'html.parser')
    tables = soup.find_all('table')
    if(len(tables)  > 0):
        table = tables[0]
        table
        rows = table.find_all('tr')
        data = [] 
        i = 0
        for row in rows:
            j = 0
            columns = row.find_all('td')
            row = []
            for column in columns:
                text = column.string
                row.append(text)
            data.append(row)
        rain = float(data[1][1])
        snow = float(data[2][1])
        elevation = int(data[9][1])
    countyRain.append(rain)
    countySnow.append(snow)
    countyElevation.append(elevation)
    


```

```python
print(countyRain)
print(countySnow) 
print(countyElevation)
```

```python
vaData['avgRain'] = countyRain 
vaData['avgSnow'] = countySnow 
vaData['countyElevation'] = countyElevation
vaData
```

```python
for index, row in vaData.iterrows():
    if row.avgRain == -1:
            vaData.loc[index, 'avgRain'] = vaData.avgRain.mean()
    if row.avgSnow == -1:
            vaData.loc[index, 'avgSnow'] = vaData.avgSnow.mean
    if row.countyElevation == -1:
            vaData.loc[index, 'countyElevation'] = vaData.countyElevation.mean()
vaData
        
        
```
